## circle mask

```python
# mask_from_csv_final_working.py
# Works with ANY image sizes in the same batch
# ~5000–8000 images/sec on RTX 4090, ~1500–3000 on RTX 3060

import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader
from pathlib import Path
import cv2
from tqdm import tqdm
```

```python
# ========================= CONFIG =========================
CSV_PATH = "../../dataset/milk_data.csv"
IMAGE_COLUMN = "NewFileName"
INPUT_ROOT = Path("../../dataset")
OUTPUT_FOLDER = Path("masked_images")
OUTPUT_FOLDER.mkdir(parents=True, exist_ok=True)

RADIUS_RATIO = 0.49
BATCH_SIZE = 32  # 256 or 512 if you have ≥16GB VRAM
NUM_WORKERS = 2
# =========================================================


class CSVDataset(Dataset):
    def __init__(self, csv_path, image_column="path", root=None):
        df = pd.read_csv(csv_path)
        self.paths = df[image_column].tolist()
        self.root = Path(root) if root else None

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, idx):
        path = self.root / self.paths[idx]
        img = cv2.imread(str(path))
        if img is None:
            raise FileNotFoundError(f"Cannot read: {path}")
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = (
            torch.from_numpy(img).permute(2, 0, 1).float() / 255.0
        )  # (C,H,W) float32 0-1
        return img, path.name


def apply_dermatoscope_mask_clean(
    batch: torch.Tensor,
    radius_from_weidth: float = 0.498,  # 0.49–0.502 → perfect top/bottom touch
    softness: float = 0.80,  # 0.7–1.1, higher = wider & gentler falloff
    strength: float = 0.89,  # 0.0 = no mask, 1.0 = full classic darkening
    center_x_offset: float = 0.0,  # ±0.1 to shift horizontally if needed
) -> torch.Tensor:
    """
    Real dermatoscope mask — NO artificial brightening, NO warm tint.
    Only soft circular darkening (exactly like hospital devices).
    Perfect for datasets and models that need authentic illumination.
    """

    def mask_single(img: torch.Tensor) -> torch.Tensor:
        C, H, W = img.shape[-3:]
        device = img.device
        dtype = torch.float32

        # Center
        cx = (W - 1) * 0.5 * (1.0 + center_x_offset)
        cy = (H - 1) * 0.5

        # Radius from height only → circle always touches top & bottom
        radius = radius_from_weidth * W

        y, x = torch.meshgrid(
            torch.arange(H, device=device, dtype=dtype),
            torch.arange(W, device=device, dtype=dtype),
            indexing="ij",
        )

        dist = torch.sqrt((x - cx) ** 2 + (y - cy) ** 2)

        # Pure raised-cosine falloff — no brightening at all
        t = torch.clamp(dist / radius, 0.0, 1.0)
        mask = 0.5 * (1.0 + torch.cos(t * torch.pi))  # 1.0 at center → 0.0 at edge
        mask = mask**softness  # controls softness

        # Strength control: 0.0 = flat image, 1.0 = full vignetting
        final_mask = 1.0 - strength * (1.0 - mask)

        return img * final_mask.unsqueeze(0)  # only darken, never brighten

    return torch.vmap(mask_single)(batch)

```

```python

# ========================= MAIN =========================
dataset = CSVDataset(CSV_PATH, IMAGE_COLUMN, INPUT_ROOT)

dataloader = DataLoader(
    dataset,
    batch_size=BATCH_SIZE,
    num_workers=NUM_WORKERS,
    pin_memory=True,
    prefetch_factor=2,
    persistent_workers=True,
)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Running on {device} | {len(dataset)} images")

for batch_imgs, batch_names in tqdm(dataloader, desc="Circular masking"):
    batch_imgs = batch_imgs.to(device, non_blocking=True)

    # This one line does everything — works with mixed sizes, no loops!
    masked = apply_dermatoscope_mask_clean(batch_imgs, RADIUS_RATIO, 0.4, 0.7, 0)
    masked = apply_dermatoscope_mask_clean(masked, RADIUS_RATIO * 1.1, 0.4, 0.7, 0)
    # Convert to uint8 numpy for saving
    masked = (masked * 255).clamp(0, 255).byte()
    masked_np = masked.permute(0, 2, 3, 1).cpu().numpy()  # (B,H,W,C)

    for img_np, name in zip(masked_np, batch_names):
        out_path = OUTPUT_FOLDER / name
        cv2.imwrite(str(out_path), cv2.cvtColor(img_np, cv2.COLOR_RGB2BGR))

print(f"\nFinished! {len(dataset)} images saved to → {OUTPUT_FOLDER.resolve()}")
```

## argumentation

```python
CSV_PATH = "../../dataset/milk_arg_data.csv"
OUTPUT_FOLDER = "arged"

INPUT_ROOT = Path("../../dataset")
OUTPUT_FOLDER = Path("arged_image")
OUTPUT_FOLDER.mkdir(parents=True, exist_ok=True)

dataset_arg = CSVDataset(CSV_PATH, IMAGE_COLUMN, INPUT_ROOT)
dataloader = DataLoader(
    dataset_arg,
    batch_size=BATCH_SIZE,
    num_workers=NUM_WORKERS,
    pin_memory=True,
    prefetch_factor=2,
    persistent_workers=True,
)


def aug_flip_h(image: np.ndarray) -> np.ndarray:
    return cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)


def aug_rand_crop(image: np.ndarray, crop_ratio: float = 0.9) -> np.ndarray:
    h, w = image.shape[:2]
    new_h, new_w = int(h * crop_ratio), int(w * crop_ratio)
    y = np.random.randint(0, h - new_h + 1)
    x = np.random.randint(0, w - new_w + 1)
    return image[y : y + new_h, x : x + new_w]


def aug_brightness(image: np.ndarray, beta: int = 30) -> np.ndarray:
    return cv2.convertScaleAbs(image, beta=beta)


def aug_gaussian_blur(image: np.ndarray, k: int = 5) -> np.ndarray:
    return cv2.GaussianBlur(image, (k, k), 0)


def aug_color_jitter(image: np.ndarray, hue_shift: int = 10) -> np.ndarray:
    img_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    img_hsv[:, :, 0] = (img_hsv[:, :, 0].astype(int) + hue_shift) % 180
    return cv2.cvtColor(img_hsv, cv2.COLOR_HSV2BGR)


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Running on {device} | {len(dataset_arg)} images")

for batch_imgs, batch_names in tqdm(dataloader, desc="Circular masking"):
    batch_imgs = batch_imgs.to(device, non_blocking=True)

    for img_np, name in zip(masked_np, batch_names):
        out_path = OUTPUT_FOLDER / name
        cv2.imwrite(str(out_path), cv2.cvtColor(img_np, cv2.COLOR_RGB2BGR))

    break

print(f"\nFinished! {len(dataset_arg)} images saved to → {OUTPUT_FOLDER.resolve()}")
```
