# Dataprocessing

## dataset merge

```python
csv_files = glob.glob(os.path.join(DATASET_BASE_DIR, "**/*.csv"), recursive=True)

df_list = []
for f in csv_files:
    if os.path.getsize(f) > 0:  # check file is not empty
        tmp = pd.read_csv(f)
        folder_nmae = f.split("/")[-2]
        tmp["NewFileName"] = folder_nmae + "/" + tmp["NewFileName"]
        df_list.append(tmp)
    else:
        print(f"Skipping empty file: {f}")

df = pd.concat(df_list, ignore_index=True)

df = df[["NewFileName", "Class", "Age", "Location", "Gender"]]
df.to_csv(os.path.join(DATASET_BASE_DIR, "data.csv"), index=False)
```

```python
df = pd.read_csv('../../dataset/data.csv')
class_list = df["Class"].unique()
df_by_class = {}
for class_item in class_list:
    df_temp = df[df["Class"] == class_item]
    if not os.path.exists(f"/~/cancer-ai/manager/dataset/csv_class/{class_item}"):
        os.mkdir(f"../../dataset/csv_class/{class_item}")
    df_temp.to_csv(f"../../dataset/csv_class/{class_item}/labels.csv", index=False)
    df_by_class[class_item] = df_temp
```
